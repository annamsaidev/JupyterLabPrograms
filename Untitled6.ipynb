```python
s = input()
def alpnum(s):
    for i in s:
        if i.isalnum():
            return 'true'
        else:
            return 'false'
        
def alpha(s):
    for i in s:
        if i.isalpha():
            return 'true'
        else:
            return 'flase'

def digit(s):
    for i in s:
        if i.isdigit():
            return 'ture'
        else:
            return 'flase'

def lowercase(s):
    for i in s:
        if i.islower():
            return 'ture'
        else:
            return 'false'
        
def uppercase(s):
    for i in s:
        if i.isupper():
            return 'true'
        else:
            return 'false'
print(alpnum(s))
print(alpha(s))
print(digit(s))
print(lowercase(s))
print(uppercase(s))
```
